```python
import numpy as np
import pandas as pd

from tldextract import extract

import warnings

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer,  TfidfTransformer

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df_imd = pd.read_csv('imdbreviewsptbr_20240514190551.csv')
df_imd.head()
```

```python
# informações do dataframe
df_imd.describe()
```

```python
print(df_imd['sentiment'].value_counts())
print(round(df_imd['sentiment'].value_counts(normalize=True)*100, 2))
```

```python
encoder = LabelEncoder()
print(df_imd['sentiment'].value_counts())
df_imd['Codedsentiment'] = encoder.fit_transform(df_imd['sentiment'])
print(df_imd['Codedsentiment'].value_counts())
```

```python
df_imd
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(df_imd['text_pt'], df_imd['Codedsentiment'], 
                                                        test_size=.2, random_state=42)
```

```python
vectorizer = CountVectorizer(ngram_range=(1,3), min_df = 0.0, max_df=1)
x_treino_vetorizado = vectorizer.fit_transform(x_treino)
x_test_vetorizado = vectorizer.transform(x_teste)
```

```python
tfidf_transformer = TfidfTransformer()
x_treino_tfidf = tfidf_transformer.fit_transform(x_treino_vetorizado)
x_teste_tfidf = tfidf_transformer.transform(x_test_vetorizado)
```

```python
classificador = MultinomialNB()
classificador.fit(x_treino_tfidf, y_treino)
```

```python
y_pred = classificador.predict(x_teste_tfidf)
```

```python
print(f'Acurácia do modelo Linear: {round(accuracy_score(y_teste, y_pred)*100,2)}%')
```
